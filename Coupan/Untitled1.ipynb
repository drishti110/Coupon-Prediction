```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Define the number of records
num_records = 78368

# Generate campaign_id in the range of 20 to 29
campaign_id = np.random.randint(20, 30, size=num_records)

# Generate campaign_type as random choices of 'Y' or 'N'
campaign_type = np.random.choice(['Y', 'N'], size=num_records)

# Generate random start dates and end dates within a range
start_dates = []
end_dates = []

start_date = datetime(2013, 1, 1)
end_date = datetime(2013, 12, 31)

for _ in range(num_records):
    start = start_date + timedelta(days=np.random.randint(0, (end_date - start_date).days))
    end = start + timedelta(days=np.random.randint(1, 60))  # Campaign lasts between 1 to 60 days
    start_dates.append(start.strftime('%d/%m/%y'))
    end_dates.append(end.strftime('%d/%m/%y'))

# Create the DataFrame
campaign_data = pd.DataFrame({
    'campaign_id': campaign_id,
    'campaign_type': campaign_type,
    'start_date': start_dates,
    'end_date': end_dates
})

# Save to CSV
campaign_data.to_csv('campaign_data.csv', index=False)
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Define the number of records
num_records = 78368

# Generate random dates within a range
start_date = datetime(2023, 1, 1)
end_date = datetime(2023, 12, 31)
date_range = end_date - start_date
dates = [start_date + timedelta(days=np.random.randint(0, date_range.days)) for _ in range(num_records)]

# Generate customer_id in the range of 1 to 3000
customer_id = np.random.randint(1, 3001, size=num_records)

# Generate item_id in the range of 10 to 99
item_id = np.random.randint(10, 100, size=num_records)

# Generate quantity as random integers from 1 to 10
quantity = np.random.randint(1, 11, size=num_records)

# Generate selling_price as random floats between 10 and 1000
selling_price = np.random.uniform(10, 1000, size=num_records).round(2)

# Generate other_discount as random floats between 0 and 100
other_discount = np.random.uniform(0, 100, size=num_records).round(2)

# Generate coupon_discount as random floats between 0 and 50
coupon_discount = np.random.uniform(0, 50, size=num_records).round(2)

# Create the DataFrame
customer_transaction_data = pd.DataFrame({
    'date': dates,
    'customer_id': customer_id,
    'item_id': item_id,
    'quantity': quantity,
    'selling_price': selling_price,
    'other_discount': other_discount,
    'coupon_discount': coupon_discount
})

# Save to CSV
customer_transaction_data.to_csv('customer_transaction_data.csv', index=False)
```

```python
# definning a function to map all the table values into one table
def add_column(column_key,column_name,table_name):
    if table_name[column_key].duplicated().any():
    # Handle duplicates (e.g., drop duplicates or use a different mapping approach)
        print("Warning: Duplicate values found in column_key. Adjusting approach.")
    # Example: Drop duplicates
        table_name = table_name.drop_duplicates(subset=[column_key])

# Proceed with mapping
    #data[column_name] = data[column_key].map(table_name.set_index(column_key)[column_name])
    data[column_name]= data[column_key].map(table_name.set_index(column_key)[column_name])

# defining a function to map values explicitly from one table to other
def add_column_explicit(column_key,column_name,table_name):
    mapping = dict(table_name[[column_key, column_name]].values)
    data[column_name] = data[column_key].map(mapping)
```
